**Forecasting with FB prophet Algorithm**

```python
# Forecasting -> Time series data --> depend upon the date and time for predict the future values
# regression --> used to predict the numerical values
```

```python
# converting the date from DD/MM/YYYY TO YYYY/MM/DD
date_info=pd.to_datetime(air_quality_data["Date"], format="%d/%m/%Y")
print(date_info)
```

```python
# converting the Time from HH.MM.SS TO HH:MM:SS
time_info=air_quality_data["Time"]
print(time_info)
```

```python
time_info=time_info.apply(lambda x:x.replace(".",":"))
```

```python
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
# Combining two series to a panda dataframe
date_time=pd.concat([date_info, time_info],axis=1)
```

```python
date_time.head()
```

```python
from prophet import Prophet
```

```python
# Combining date and time -> YYYY-MM-DD HH:MM:SS
date_time["ds"]=date_time["Date"].astype(str)+" "+date_time["Time"].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

```python
# we have to convert ds from object datatype to date time format
```

```python
data = pd.DataFrame()
```

```python
data["ds"]=pd.to_datetime(date_time["ds"])
```

```python
data.head()
```

```python
# RH -> Relative humidity
# data["y"]=air_quality_data["AH"]
data["y"]=air_quality_data["RH"]
```

```python
data.head()
```

```python
# training the prophet model

model = Prophet()
model.fit(data)
```

```python
data.tail()
```

```python
# H stands for "hour", which means that the future dataframe will have timestamps that are spaced one hour apart.
future=model.make_future_dataframe(periods=365)
future.tail()
```

```python
# yhat: This represents the point estimate of the predicted value
# yhat_lower: This represents the lower bound of the prediction
# yhat_upper: This represents the upper bound of the prediction

forecast=model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
